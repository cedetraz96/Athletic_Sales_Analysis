```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_athletic_sales = pd.concat([athletic_sales_2020, athletic_sales_2021], axis="rows", join="inner")
joined_athletic_sales.head()

joined_athletic_sales = joined_athletic_sales.reset_index(drop=True)
joined_athletic_sales.head()
```

```python
# Check if any values are null.
joined_athletic_sales = joined_athletic_sales.dropna()
```

```python
# Check the data type of each column
joined_athletic_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_athletic_sales = joined_athletic_sales.rename(columns={"invoice_date": "datetime"})
joined_athletic_sales.head()
```

```python
# Confirm that the "invoice_date" data type has been changed.
joined_athletic_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = joined_athletic_sales.groupby(["region", "state", "city"])["units_sold"].sum()
total_products_sold = joined_athletic_sales.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_products_sold.head(5)                           
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold_sum = pd.pivot_table(joined_athletic_sales, 
                                                index=["region", "state", "city"], 
                                                values="units_sold", 
                                                aggfunc="sum")

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold_sum = total_products_sold_sum.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_products_sold_sum.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = joined_athletic_sales.groupby(["region", "state", "city"])["total_sales"].sum()
total_sales = joined_athletic_sales.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
total_sales.head(5)  
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_sum = pd.pivot_table(joined_athletic_sales,
                                 index=["region", "state", "city"],
                                 values="total_sales",
                                 aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_sum = total_sales_sum.rename(columns={"total_sales": "Total_Sales"})

# Show the top 5 results.
total_sales_sum.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = joined_athletic_sales.groupby(["retailer", "region", "state", "city"])["total_sales"].sum()
total_sales = joined_athletic_sales.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
total_sales.head(5)  
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_sum = pd.pivot_table(joined_athletic_sales,
                                 index=["retailer", "region", "state", "city"],
                                 values="total_sales",
                                 aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_sum = total_sales_sum.rename(columns={"total_sales": "Total_Sales"})

# Show the top 5 results.
total_sales_sum.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_sales = joined_athletic_sales[joined_athletic_sales["product"]  == "Women's Athletic Footwear"]
womens_athletic_footwear_sales.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_sales_womens_athletic_footwear = womens_athletic_footwear_sales.groupby(["retailer", "region", "state", "city"])["total_sales"].sum()
total_sales_womens_athletic_footwear = womens_athletic_footwear_sales.rename(columns={"total_sales": "Total_Sales"})
# Show the top 5 results.
total_sales_womens_athletic_footwear.head(5)  
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_womens_athletic_footwear_sum = pd.pivot_table(joined_athletic_sales,
                                 index=["retailer", "region", "state", "city"],
                                 values="total_sales",
                                 aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_womens_athletic_footwear_sum = total_sales_womens_athletic_footwear_sum.rename(columns={"total_sales": "Total_Sales"})

# Show the top 5 results.
total_sales_womens_athletic_footwear_sum.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_total_sales_womens_athletic_footwear = pd.pivot_table(womens_athletic_footwear_sales, 
                                                                  index="datetime", 
                                                                  values="total_sales")

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_total_sales_womens_athletic_footwear = pivot_table_total_sales_womens_athletic_footwear.rename(columns={"total_sales": "Total_Sales"})

# Show the table.
pivot_table_total_sales_womens_athletic_footwear.head(5)
```

```python
# Resample the pivot table into daily bins and get the total sales for each day
daily_total_sales_womens_athletic_footwear = pivot_table_total_sales_womens_athletic_footwear.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_table_total_sales_womens_athletic_footwear.sort_values(by="Total_Sales", ascending=False).head()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_total_sales_womens_athletic_footwear = pivot_table_total_sales_womens_athletic_footwear.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_table_total_sales_womens_athletic_footwear.sort_values(by="Total_Sales", ascending=False).head()
```
